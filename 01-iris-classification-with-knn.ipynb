## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">لود کتابخانه‌های مورد استفاده</div>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">لود مجموعه داده (dataset)</div>

```python
iris = load_iris(as_frame=True)
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(iris.target_names)
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">تعریف مدل (model definition)</div>

```python
model = KNeighborsClassifier(n_neighbors=3)
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">آموزش مدل (model training)</div>

```python
model.fit(X_train, y_train)
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">پیش بینی (prediction)</div>

```python
y_pred = model.predict(X_test)
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ارزیابی (evaluation)</div>

```python
print("accuracy:",accuracy_score(y_test, y_pred))
```
